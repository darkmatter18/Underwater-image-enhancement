## Approach
- We are using Cycle GAN to enhance the underwater images.
- CycleGAN is a technique that involves the automatic training of image-to-image translation models without paired examples.
- We have 2 domain of images, I) Underwater Images and II) Enhanced underwater images and clear images.
- The models are trained in an unsupervised manner using a collection of images from the source and target domain that do not need to be related in any way.
- Our target is to train the GAN to be able to translate non-enhanced underwater image to enhanced underwater image by removing the common underwater artifacts from the image.

## Datasets Used
- The EUVP dataset
    - The EUVP (Enhancing Underwater Visual Perception) dataset contains separate sets of paired and unpaired image samples of poor and good perceptual quality to facilitate supervised training of underwater image enhancement models. 
    - Link: http://irvlab.cs.umn.edu/resources/euvp-dataset
