```python
import pickle
import matplotlib.pyplot as plt
```

```python
filename = './models_uwie_v181_loss_stats.pkl'
```

```python
with open(filename, 'rb') as f:
    losses = pickle.load(f)
```

```python
plt.figure(figsize=(15,7))
plt.plot(losses['loss_idt_A'])
plt.plot(losses['loss_idt_B'])
```

```python
plt.figure(figsize=(15,7))
plt.plot(losses['loss_D_A'], label="D_A")
plt.plot(losses['loss_D_B'], label="D_B")
plt.legend()
plt.xlabel('# iters')
plt.ylabel('Loss Value')
plt.ylim((0, 0.5))
plt.title("Discriminator Loss")
```

```python
plt.figure(figsize=(15,7))
plt.plot(losses['loss_G_AtoB'], label="G_AtoB")
plt.plot(losses['loss_G_BtoA'], label="G_BtoA")
plt.legend()
plt.xlabel('# iters')
plt.ylabel('Loss Value')
plt.ylim((0, 1.4))
plt.title("Generator Loss")
```

```python
plt.figure(figsize=(15,7))
plt.plot(losses['cycle_loss_A'], label="Cycle_A")
plt.plot(losses['cycle_loss_B'], label="Cycle_B")
plt.legend()
plt.xlabel('# iters')
plt.ylabel('Loss Value')
plt.ylim((0, 3.0))
plt.title("Cycle Loss")
```
