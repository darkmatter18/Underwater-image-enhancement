```python
import pickle
import matplotlib.pyplot as plt
```

```python
columns = list(range(26, 82))
ranges = [0,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1]

pkl_file_paths = list()
for f_name in columns:
    pkl_file_paths.append(f"output/{f_name}_data.pkl")

final_matrix = list()
for pkl_file_path in pkl_file_paths:
    with open(pkl_file_path, "rb") as pkl_file:
        pkl_file_data = pickle.load(pkl_file)
        final_matrix.append(pkl_file_data.get("ssim"))
```

```python
import numpy as np

data_matrix = np.matrix(final_matrix).transpose()

print(data_matrix)
```

```python
import pandas as pd

df = pd.DataFrame(data_matrix, columns = columns)
display(df)
```

```python
df.mean().plot(figsize=(16,8))
```

```python
df.median().plot(figsize=(16,8))
```

```python
df[80].value_counts(bins=20, sort=False)
```

```python
matrix = np.array([df[column].groupby(pd.cut(df[column], ranges)).count().values for column in df.columns.values]).transpose()
out = pd.DataFrame(matrix, columns= df.columns.values)
```

```python
out.plot(figsize=(20,20))
```

```python
out.plot(figsize=(20,20), kind="hist")
```
