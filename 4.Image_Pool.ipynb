## Underwater image enhancement model with PyTorch

This notebook contains the demo of `Image Pool`


```python
from data import create_dataset
from utils.image_pool import ImagePool
import matplotlib.pyplot as plt
```

```python
class Opt:
    batch_size=10
    checkpoints_dir='./checkpoints'
    crop_size=256
    dataroot='./Dataset/EUVP Dataset/Paired/underwater_dark'
    direction='AtoB'
    flip=False
    input_nc=3
    load_size=286
    max_dataset_size=float('inf')
    name='experiment_name'
    num_threads=0
    output_nc=3
    phase='train'
    preprocess='resize_and_crop'
    serial_batches=False
```

```python
data = create_dataset(Opt())
```

```python
i = iter(data)
```

```python
d = next(i)['A']
```

```python
imagepool = ImagePool(5)
```

```python
out = imagepool.query(d)
out.size()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=10, figsize=(20,5))
for index in range(len(d)):
    ax[0,index].imshow(d[index].permute(1, 2, 0).numpy())

for index in range(len(out)):
    ax[1,index].imshow(out[index].permute(1, 2, 0).numpy())

plt.show(block=True)
```
